### Comp Bio Sheet 3 - Question 4
#### (b)

```julia
x₀ = 1
p₀ = 0
m = 1
ω = 1
k = m * ω^2

n_steps = 10000

x_exact(t) = x₀ * cos(ω * t) + p₀ * sin(ω * t)
p_exact(t) = -ω * x₀ * sin(ω * t) + p₀ * cos(ω * t) / m

F(x) = -k * x;
```

```julia
function integration_step(x, p, f, Δt, F)
    p += Δt * f / 2
    x += Δt * p / m 
    f = F(x)
    p += Δt * f / 2
    return x, p, f
end;
```

```julia
# Part (i)

x = x₀
p = p₀
f = F(x)
Δt = 0.01

xs_simul = zeros(n_steps)
ps_simul = zeros(n_steps)

for i in 1:n_steps
    x, p, f = integration_step(x, p, f, Δt, F)
    xs_simul[i] = x
    ps_simul[i] = p
end
```

```julia
using Plots 


t_steps = Δt * (1:n_steps)

T = 2π / ω 

t_steps_scaled = t_steps / T

plot(t_steps_scaled, xs_simul, xlabel = "t/T", ylabel = "x(t)", label="")
```

```julia
# Parts (ii) and (iii)
xs_exact = x_exact.(t_steps)
ps_exact = p_exact.(t_steps)

plot(xs_simul, ps_simul, xlabel = "x", ylabel = "p", label="simulated", linewidth=0.01)
plot!(xs_exact, ps_exact, xlabel = "x", ylabel = "p", label="exact", linewidth=2)
```

```julia
# Part (iv) 

E₀= p₀^2 / (2*m) + 1/2 * m * ω^2 * x₀^2 

E(x, p) = p^2 / (2*m) + 1/2 * m * ω^2 * x^2 

E₀= E(x₀, p₀)

Es_simul = E.(xs_simul, ps_simul)

ΔE(i, Es) = 1 / i * sum(abs.((Es[1:i] .- E₀) / E₀))

ΔEs = ΔE.(1:n_steps, Ref(Es_simul));

plot(t_steps_scaled, ΔEs, xlabel = "t/T", ylabel = "x(t)", label="ΔE(t)")
```

#### (c)

```julia
λ = 0.99
F_fast(x) = λ * F(x)
F_slow(x) = (1 - λ) * F(x)

δ = 0.01
function multiple_timescale_integration_step(x, p, f_slow, f_fast, Δt, F_slow, F_fast)
    p += Δt * f_slow / 2
    for i ∈ δ:δ:Δt
        x, p, f_fast = integration_step(x, p, f_fast, δ, F_fast)
    end
    f_slow = F_slow(x)
    p += Δt * f_slow / 2
    return x, p, f_slow, f_fast
end;

t_end=100.0
energy_plot = plot(t_steps_scaled, xs_exact, xlabel = "t/T", ylabel = "x(t)", label="Analytical")
phase_plot = plot(xs_exact, ps_exact, xlabel = "x", ylabel = "p", label="Analytical")
Δts = [0.1, 0.5, 1.0]
for Δt ∈ Δts
    x = x₀
    p = p₀
    f_slow = F_slow(x)
    f_fast = F_fast(x)
    for i in 1:n_steps
        x, p, f_slow, f_fast = multiple_timescale_integration_step(x, p, f_slow, f_fast, Δt, F_slow, F_fast)
        xs_simul[i] = x
        ps_simul[i] = p
    end
    times = (Δt:Δt:t_end) / T
    n = length(times)
    plot!(energy_plot, times, xs_simul[1:n], xlabel = "t/T", ylabel = "x(t)", label="Simulation: Δt=" * string(Δt) * " δ=" * string(δ), linewidth=1.5/√Δt, alpha=0.9)
    plot!(phase_plot, xs_simul[1:n], ps_simul[1:n], xlabel = "x", ylabel = "p", label="Simulation: Δt=" * string(Δt) * " δ=" * string(δ))
end

display(energy_plot)
display(phase_plot)
```
