```julia
using Turing 
using ReverseDiff
using TorusEvol
using StatsPlots
#Turing.setadbackend(:reversediff)
Turing.setrdcache(true)
Turing.setadbackend(:forwarddiff)
Turing.setprogress!(true)

@model function bisection_sampler(p, y, z)
    t ~ Exponential(1)
    
    y ~ statdist(p)
    x ~ transdist(p, t/2, y)
    z ~ transdist(p, t/2, x)
end
```

```julia
# continuous!!

p = WrappedDiffusion(0.0, 0.0, 0.7, 0.7, 1.0, 1.0, 0.2)
x = [1.0, -2.0]
t = 0.7
y = rand(transdist(p, t/2, x))
z = rand(transdist(p, t/2, x))

model = bisection_sampler(p, y, z)
ch = sample(model, Gibbs(HMC(0.05, 10, :x), HMC(0.02, 5, :t)), 3000)
plot(ch, fontfamily="JuliaMono")
```

```julia
# discrete!!

x = 10
p = WAG_SubstitutionProcess

t_y = 0.8
y = rand(transdist(p, t_y, x))
@info y

t_z = 0.5
z = rand(transdist(p, t_z, x))
@info z

model = bisection_sampler(p, y, z, t_y, t_z)
ch = sample(model, PG(100, :x), 100)
```
